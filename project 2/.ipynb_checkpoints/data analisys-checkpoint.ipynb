```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv("consumption_and_temperatures.csv")
```

```python
df.head()
```

```python
# Convert the timestamp feature to datetime format if it's not already in datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Transform the timestamp feature into the desired format
#df['timestamp'] = df['timestamp'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
df['day'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
df['year'] = df['timestamp'].dt.year
```

```python
df
```

```python
df_2018 = df[df['timestamp'].dt.year == 2018]
```

```python
df.dtypes
```

```python
df.isna().any()
```

```python
df.set_index('timestamp', inplace=True)
```

```python
df.head()
```

```python
df['NO1_temperature'] = df['NO1_temperature'].shift(fill_value=0)
df
```

```python
#1 year plot
```

```python
df['NO1_consumption'].plot(xlim=['2018-01-01','2019-01-01 00:00'], figsize=(12,4)) 
df['NO2_consumption'].plot(xlim=['2018-01-01','2019-01-01 00:00'], figsize=(12,4)) 
df['NO3_consumption'].plot(xlim=['2018-01-01','2019-01-01 00:00'], figsize=(12,4)) 
df['NO4_consumption'].plot(xlim=['2018-01-01','2019-01-01 00:00'], figsize=(12,4)) 
df['NO5_consumption'].plot(xlim=['2018-01-01','2019-01-01 00:00'], figsize=(12,4)) 

plt.legend()

# Optionally, you can add labels and title
plt.xlabel('Date')
plt.ylabel('Consumption')
plt.title('Consumption Trends')

plt.show()
```
